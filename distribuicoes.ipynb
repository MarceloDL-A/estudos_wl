```python
import numpy as np
import matplotlib.pyplot as plt
```

# Uniforme

```python
u = np.random.uniform(5, 30, 10000)

plt.hist(u, bins = 50, density= True)
plt.title("Distribuição Uniforme")
plt.xlabel("Eventos")
plt.ylabel("Probabilidade")
plt.show()
```

# Normal

```python
# media 
# desvio
media = 0
desvio_padrao = 5
n = np.random.normal(media, desvio_padrao, 10000)

plt.hist(n, bins = 50, density= True)
plt.title("Distribuição Normal")
plt.xlabel("Valor")
plt.ylabe("Frequência")
```

# Binomial

```python
# Eventos
tentativas = 20
chance_de_sucesso = 0.05

b = np.random.binomial(tentativas, chance_de_sucesso, 10000)

plt.figure(figsize = (10, 5))
plt.hist(b, bins = range(tentativas + 2),
         align = "left",
         density = True)
plt.xticks(range(-1, tentativas + 2, 1))
plt.title("Distribuição binomial")
plt.xlabel("Número de sucessos")
plt.ylabel("Probabilidade")
plt.show()
```

# Poisson

```python
# lambda: média de eventos no tempo
lambda_ = 30

p = np.random.poisson(lambda_, 1000)

plt.hist(p, bins = 50, density = True)
plt.xlabel("Eventos")
plt.ylabel("Probabilidade")
plt.show()
```

```python
# Exponencial
```

```python
taxa = 5

exp = np.random.exponential(5, 1000)

plt.figure(figsize = (10, 5))
plt.hist(exp, bins = 30, density = True)
plt.title("Distribuição exponencial")
plt.xlabel("Itervalo de tempo")
plt.ylabel("Probabelidade")
plt.xticks(range(0, 35, 2))
plt.show()




```

```python
# Geometrica
```

```python
chance = 0.3

geom = np.random.geometric(chance, 1000)
plt.hist(geom, bins = max(geom) - min(geom), density = True)
plt.title("Densidade de distribuição geométrica")
plt.xlabel("Número de eventos até o sucesso")
plt.ylabel("Probabilidade")
plt.show()
```
