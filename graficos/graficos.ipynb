```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Configurando o estilo do Seaborn
sns.set(style="whitegrid")

# Gerando dados com distribuição geométrica
chance = 0.3
geom = np.random.geometric(chance, 1000)

# Criando o histograma com densidade usando Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(geom, bins=max(geom) - min(geom), stat='density', kde=False)

# Adicionando títulos e rótulos
plt.title("Densidade de Distribuição Geométrica")
plt.xlabel("Número de Eventos até o Sucesso")
plt.ylabel("Probabilidade")

# Exibindo o gráfico
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Configurando o estilo do Seaborn
sns.set(style="whitegrid")

# Gerando dados fictícios
np.random.seed(10)  # Para reprodutibilidade dos dados
grupos = ['Grupo A', 'Grupo B', 'Grupo C', 'Grupo D']
dados = np.random.normal(loc=100, scale=20, size=(200, 4))  # 200 pontuações para cada grupo

# Convertendo os dados para um formato adequado para o Seaborn
dados_melt = pd.melt(pd.DataFrame(dados, columns=grupos), var_name='Grupo', value_name='Pontuação')

# Criando o Box Plot
plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(x='Grupo', y='Pontuação', data=dados_melt)

# Adicionando títulos e rótulos
plt.title('Distribuição de Pontuações por Grupo')
plt.xlabel('Grupo')
plt.ylabel('Pontuação')

# Exibindo o gráfico
plt.show()
```
