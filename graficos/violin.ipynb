Gráfico de Violino para Distribuição das Características das Íris

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Carregando o dataset Iris
iris = sns.load_dataset('iris')
```

```python
# Agrupando os dados por espécie e calculando as estatísticas relevantes para o gráfico de violino
estatisticas_por_especie = iris.groupby('species')['petal_length'].describe()

# Selecionando as estatísticas de interesse
estatisticas_interesse = estatisticas_por_especie[['25%', '50%', '75%', 'min', 'max']]

print("petal_length\n", estatisticas_interesse)


# Criando gráficos de violino para visualizar a distribuição das características das íris
plt.figure(figsize=(12, 6))
sns.violinplot(x="species", y="petal_length", data=iris)
plt.title('Gráfico de Violino: Comprimento das Pétalas por Espécie')
plt.show()
```
