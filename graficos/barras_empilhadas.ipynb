```python
import matplotlib.pyplot as plt
import numpy as np

# Simulando um conjunto de dados de estudos
categorias = ['Matemática', 'Ciência', 'História', 'Línguas', 'Artes']
estudante_A = np.array([15, 20, 10, 30, 25])
estudante_B = np.array([10, 25, 5, 20, 30])
estudante_C = np.array([20, 15, 15, 10, 20])

# Criando o gráfico de barras empilhadas
plt.figure(figsize=(10, 7))

# Empilhando os valores para cada estudante
plt.bar(categorias, estudante_A, color='r', label='Estudante A')
plt.bar(categorias, estudante_B, bottom=estudante_A, color='b', label='Estudante B')
plt.bar(categorias, estudante_C, bottom=estudante_A+estudante_B, color='g', label='Estudante C')

# Adicionando detalhes ao gráfico
plt.xlabel("Disciplinas")
plt.ylabel("Horas Estudadas")
plt.title("Horas de Estudo por Disciplina e Estudante")
plt.legend()

# Mostrando o gráfico
plt.show()
```
