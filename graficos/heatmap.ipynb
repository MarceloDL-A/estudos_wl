 Heatmap da Correlação entre Características das Íris


- `corr = iris.corr()`: Calcula a matriz de correlação de Pearson para as variáveis numéricas do dataset `Iris`.
- `sns.heatmap()`: Cria o heatmap a partir da matriz de correlação. Os argumentos utilizados são:
  - `corr`: A matriz de correlação calculada.
  - `annot=True`: Adiciona anotações para cada célula do heatmap com os coeficientes de correlação.
  - `cmap='coolwarm'`: Define o mapa de cores para o heatmap, onde cores quentes (como vermelho) indicam uma correlação positiva e cores frias (como azul) indicam uma correlação negativa.
  - `fmt=".2f"`: Formata os números anotados para duas casas decimais.
  - `center=0`: Define o ponto central do mapa de cores em 0 para que correlações negativas e positivas sejam facilmente diferenciadas.
  - `square=True`: Garante que as células do heatmap sejam quadradas.
  - `linewidths=.5`: Define a largura das linhas que separam as células.
  - `cbar_kws={"shrink": .5}`: Ajusta a barra de cores do heatmap.

Este heatmap fornece uma visualização intuitiva das correlações entre as características das flores de íris, ajudando a identificar relações potenciais que podem ser importantes para análises mais detalhadas, como a construção de modelos de machine learning.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando o dataset Iris
iris = sns.load_dataset('iris')

# Selecionando apenas as colunas numéricas (excluindo a coluna 'species')
iris_numerico = iris.select_dtypes(include=['float64', 'int64'])

# Calculando a matriz de correlação apenas para as colunas numéricas
corr = iris_numerico.corr()

# Criando o heatmap para visualizar a matriz de correlação
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Heatmap de Correlação entre Características das Íris')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Carregando o dataset Iris
iris = sns.load_dataset('iris')

# Selecionando apenas as colunas numéricas (excluindo a coluna 'species')
iris_numerico = iris.select_dtypes(include=['float64', 'int64'])

# Calculando a matriz de correlação apenas para as colunas numéricas
corr = iris_numerico.corr()

# Criando uma máscara para ocultar a parte superior da matriz de correlação
mask = np.triu(np.ones_like(corr, dtype=bool))

# Criando o heatmap para visualizar a matriz de correlação
plt.figure(figsize=(8, 6))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', fmt=".2f", center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Heatmap de Correlação entre Características das Íris (Metade Inferior)')
plt.show()
```
