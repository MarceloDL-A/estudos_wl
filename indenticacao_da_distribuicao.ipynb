```python
from fitter import Fitter
import numpy as np
```

```python
data = np.random.normal(35, 3, 10000)
```

```python
# Lista expandida de distribuições conhecidas para testar, incluindo t de Student, uniforme e qui-quadrado
distributions = [
    'norm',    # Normal
    'binom',   # Binomial (note que binom precisa de argumentos adicionais como n e p)
    'poisson', # Poisson
    'expon',   # Exponencial
    'gamma',   # Gamma
    't',       # t de Student
    'uniform', # Uniforme
    'chi2'     # Qui-quadrado (note que chi2 precisa do argumento df - graus de liberdade)
]

# Criar uma instância do Fitter com a lista de distribuições
f = Fitter(data, distributions=distributions)
f.fit()
```

```python
f.summary(Nbest=10)
```
