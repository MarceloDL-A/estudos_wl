```python
import pandas as pd
import numpy as np

# Definindo a semente aleatória para reprodutibilidade
np.random.seed(42)

# Número de registros
n = 1000

# Gerando dados fictícios
data = {
    "ID": range(1, n + 1),
    "Nome": ["Jogo_" + str(i) for i in range(1, n + 1)],
    "Gênero": np.random.choice(["Ação", "Aventura", "RPG", "Esporte", "Estratégia"], size=n),
    "Plataforma": np.random.choice(["PC", "PS4", "Xbox One", "Switch"], size=n),
    "Ano de Lançamento": np.random.randint(2000, 2022, size=n),
    "Vendas Globais": np.round(np.random.uniform(0.1, 5.0, size=n), 2),
    "Avaliação dos Usuários": np.round(np.random.uniform(1, 10, size=n), 1),
    "Avaliação da Crítica": np.random.randint(1, 100, size=n),
    "Classificação Etária": np.random.choice(["E", "T", "M"], size=n)
}

# Criando o DataFrame
df_jogos = pd.DataFrame(data)

# Visualizando as primeiras linhas do DataFrame
df_jogos.head()
```

```python
# Estatísticas descritivas
df_jogos.describe().round(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração do estilo dos gráficos
sns.set(style="whitegrid")

# Histograma das Avaliações dos Usuários
plt.figure(figsize=(10, 6))
sns.histplot(df_jogos['Avaliação dos Usuários'], kde=True, bins=20)
plt.title('Distribuição das Avaliações dos Usuários')
plt.xlabel('Avaliação')
plt.ylabel('Frequência')
plt.show()

# Gráfico de barras para Vendas Globais por Gênero
plt.figure(figsize=(10, 6))
sns.barplot(x='Gênero', y='Vendas Globais', data=df_jogos, estimator=sum, ci=None)
plt.title('Vendas Globais por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Vendas Globais (em milhões)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico de dispersão entre Avaliação dos Usuários e Avaliação da Crítica
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Avaliação da Crítica', y='Avaliação dos Usuários', data=df_jogos)
plt.title('Avaliação dos Usuários vs Avaliação da Crítica')
plt.xlabel('Avaliação da Crítica')
plt.ylabel('Avaliação dos Usuários')
plt.show()
```
