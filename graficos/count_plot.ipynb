```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando o dataset Iris
iris = sns.load_dataset('iris')

# Criando um gráfico de contagem para as espécies de íris
sns.countplot(data=iris, x='species')
plt.title('Contagem das Espécies de Íris')
plt.show()
```

```python
df_titanic = sns.load_dataset('titanic')

plt.figure(figsize=(10, 6))
sns.countplot(data=df_titanic, x='class')
plt.title('Distribuição dos Passageiros por Classe')
plt.xlabel('Classe de Passageiro')
plt.ylabel('Contagem')
plt.show()
```
