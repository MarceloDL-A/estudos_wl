```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Gráfico de Linha
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.plot(x, y, label='sin(x)')
plt.title('Gráfico de Linha')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()

# Gráfico de Barras
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores = [10, 20, 15, 8]
plt.subplot(2, 2, 2)
plt.bar(categorias, valores, color='skyblue')
plt.title('Gráfico de Barras')
plt.xlabel('Categorias')
plt.ylabel('Valores')

# Gráfico de Dispersão
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.subplot(2, 2, 3)
plt.scatter(x_scatter, y_scatter, color='green')
plt.title('Gráfico de Dispersão')
plt.xlabel('x')
plt.ylabel('y')

# Histograma
dados_histograma = np.random.randn(1000)
plt.subplot(2, 2, 4)
plt.hist(dados_histograma, bins=30, color='orange')
plt.title('Histograma')
plt.xlabel('Valor')
plt.ylabel('Frequência')

# Ajustando layout para evitar sobreposição
plt.tight_layout()

# Mostrar os gráficos
plt.show()
```
