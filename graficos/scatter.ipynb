```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Carregando o conjunto de dados Iris
df = sns.load_dataset('iris')

# Criando o grfico de disperso
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title('Relao entre Comprimento da Spala e Comprimento da Ptala por Espcie de ris')
plt.xlabel('Comprimento da Spala (cm)')
plt.ylabel('Comprimento da Ptala (cm)')
plt.legend(title='Espcie')
plt.show()
```
