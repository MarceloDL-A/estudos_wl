```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python

# Carrega dataset
df = pd.read_csv('../datasets/bank/bank-full.csv', delimiter=';')
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# Assumindo que 'df' seja o DataFrame carregado com os dados do dataset "Bank Marketing"

# SeleÃ§Ã£o de variÃ¡veis para anÃ¡lise
features = ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'duration']

# PreparaÃ§Ã£o dos dados
X = df[features]

# PrÃ©-processamento com Pipelines
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['age', 'balance', 'duration']),
        ('cat', OneHotEncoder(), ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact'])
    ])

# Processamento dos dados
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
X_processed = pipeline.fit_transform(X)

# MÃ©todo do Cotovelo
wcss = []
for i in range(1, 11):  # Testando de 1 a 10 clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_processed)
    wcss.append(kmeans.inertia_)

# Plotando o grÃ¡fico do Cotovelo
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='blue')
plt.title('MÃ©todo do Cotovelo para Determinar o NÃºmero Ã“timo de Clusters')
plt.xlabel('NÃºmero de Clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

# Definindo o nÃºmero Ã³timo de clusters baseado na anÃ¡lise visual do grÃ¡fico do cotovelo
n_clusters_optimo = 4

# Aplicando K-Means com o nÃºmero Ã³timo de clusters
kmeans_optimo = KMeans(n_clusters=n_clusters_optimo, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans_optimo.fit(X_processed)
df['cluster'] = kmeans_optimo.labels_

# AnÃ¡lise PÃ³s-ClusterizaÃ§Ã£o
# EstatÃ­sticas por cluster ajustadas para o nÃºmero correto de clusters
for i in range(n_clusters_optimo):
    print(f"\nCluster {i} EstatÃ­sticas\n")
    display(df[df['cluster'] == i].describe())

# VisualizaÃ§Ã£o da distribuiÃ§Ã£o de 'age' e 'balance' por cluster
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(x='cluster', y='age', data=df, ax=ax[0])
sns.boxplot(x='cluster', y='balance', data=df, ax=ax[1])
plt.tight_layout()
plt.show()

# AvaliaÃ§Ã£o do Modelo com Silhouette Score usando os rÃ³tulos corretos
X_transformed = pipeline.transform(X)  # Utilizando o pipeline para garantir a mesma transformaÃ§Ã£o
score = silhouette_score(X_transformed, kmeans_optimo.labels_)
print(f"\nSilhouette Score: {score:.2f}")
```

```python
# Silhouete score para diferentes nÃºmeros de clusters

from sklearn.metrics import silhouette_score

silhouette_scores = []
for i in range(2, 11):  # K-Means nÃ£o pode ser aplicado com apenas 1 cluster
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_processed)
    score = silhouette_score(X_processed, kmeans.labels_)
    silhouette_scores.append(score)

# Plotando os Silhouette Scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-', color='blue')
plt.title('Silhouette Score para Diferentes NÃºmeros de Clusters')
plt.xlabel('NÃºmero de Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

# Assumindo que 'df' seja o DataFrame carregado com os dados do dataset "Bank Marketing"

# SeleÃ§Ã£o de variÃ¡veis para anÃ¡lise
features = ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'duration']

# PreparaÃ§Ã£o dos dados
X = df[features]

# PrÃ©-processamento com Pipelines
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['age', 'balance', 'duration']),
        ('cat', OneHotEncoder(), ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact'])
    ])

# Processamento dos dados
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
X_processed = pipeline.fit_transform(X)

# Aplicando KMeans para definir os labels dos clusters
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)  # Ajuste o n_clusters conforme necessÃ¡rio
kmeans.fit(X_processed)
labels = kmeans.labels_  # RÃ³tulos dos clusters

# Aplicando PCA nos dados prÃ©-processados
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_processed)

# Visualizando os clusters com PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50, alpha=0.6)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('VisualizaÃ§Ã£o dos Clusters com PCA - Dataset do Banco')
plt.colorbar(label='Cluster')
plt.show()

# VariaÃ§Ã£o explicada por cada componente principal
print("VariaÃ§Ã£o explicada por cada componente principal:", pca.explained_variance_ratio_)
```
